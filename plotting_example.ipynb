```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re
```

```python
files = os.listdir()
r = re.compile('.*_log.tsv')
logs = list(filter(r.match, files))
log_file = logs[0]
```

```python
df = pd.read_csv(log_file, sep='\t')
```

```python
time = df.TIME
vms = df.VMS
rss = df.RSS
cpu = df['%CPU']
fd = df.FD

fig, axs = plt.subplots(2, 2, figsize=(9, 6))
axs[0, 0].plot(time, vms)
axs[0, 1].plot(time, rss)
axs[1, 0].plot(time, cpu)
axs[1, 1].plot(time, fd)

axs[0, 0].set_title('VMS, mb')
axs[0, 1].set_title('RSS, mb')
axs[1, 0].set_title('%CPU')
axs[1, 1].set_title('FD count')

axs[0, 0].set(xticklabels=[])
axs[0, 1].set(xticklabels=[])
axs[1, 0].set(xticklabels=[])
axs[1, 1].set(xticklabels=[])

fig.suptitle('Resources usage by ' + log_file.split('_log.tsv')[0])
```
